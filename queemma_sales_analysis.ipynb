```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel("Queemma_staff.xlsx")

df.head()
```

## 3. Data Cleaning

```python
# Remove empty columns
df = df.drop(columns=['Unnamed: 9','Unnamed: 10','Unnamed: 11'])

# Convert date column
df['Date'] = pd.to_datetime(df['Date'])

# Check duplicates
df = df.drop_duplicates()

df.head()
```

```python
# Validate sales calculation
df['Calculated Sales'] = df['Units Sold'] * df['Unit Price']
df[['Sales Amount','Calculated Sales']].head()
```

## 4. Exploratory Data Analysis

```python
# Total Revenue
df['Sales Amount'].sum()
```

```python
# Monthly Sales Trend
monthly_sales = df.groupby('month')['Sales Amount'].sum()

monthly_sales.plot(kind='line',title='Monthly Sales Trend')
plt.show()
```

```python
# Sales by Region
region_sales = df.groupby('Region')['Sales Amount'].sum().sort_values()

region_sales.plot(kind='bar',title='Sales by Region')
plt.show()
```

```python
# Sales by Product
product_sales = df.groupby('Product')['Sales Amount'].sum().sort_values()

product_sales.plot(kind='bar',title='Sales by Product')
plt.show()
```

```python
# Salesperson performance
salesperson_sales = df.groupby('Salesperson')['Sales Amount'].sum().sort_values()

salesperson_sales.plot(kind='bar',title='Sales by Salesperson')
plt.show()
```

## 5. Advanced Analysis

```python
# Top 5 products
df.groupby('Product')['Sales Amount'].sum().sort_values(ascending=False).head()
```

```python
# Top 5 salespeople
df.groupby('Salesperson')['Sales Amount'].sum().sort_values(ascending=False).head()
```

```python
# Region contribution
region_contribution = df.groupby('Region')['Sales Amount'].sum()
region_contribution/region_contribution.sum()*100
```
